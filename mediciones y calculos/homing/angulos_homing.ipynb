```python
import pandas as pd
```

```python
# load CSV into a dataframe
df = pd.read_csv('angulos_homing.csv')
df.head()
```

```python
# calculate mean of each angle
means = df[['angulo1', 'angulo2']].mean()
print(means)

# individual mean values
mean_angulo1 = means['angulo1']
mean_angulo2 = means['angulo2']
print(f"Mean angulo1: {mean_angulo1}")
print(f"Mean angulo2: {mean_angulo2}")
```
